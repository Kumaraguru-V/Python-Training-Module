## **3.1. Functions and its features**

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Functions in Python are blocks of reusable code that perform a specific task. They allow you to break down your program into smaller, manageable parts, making your code more organized, readable, and maintainable. Here are some key aspects of functions in Python:

* **Function Definition:**

You define a function using the `def` keyword, followed by the function name and a pair of parentheses. Any input parameters are listed within the parentheses, and the function body is indented below the definition. You can also specify a return value using the return statement.

```python
def greet(name):
    return f"Hello, {name}!"
```

* **Function Call:**

To execute a function, you call it by its name followed by parentheses. If the function takes parameters, you pass the values inside the parentheses.

```python
message = greet("Alice")
print(message)  # Output: Hello, Alice!
```

* **Parameters and Arguments:**

Parameters are placeholders for values that are passed into a function. Arguments are the actual values that are passed to the function when it is called. Functions can have multiple parameters, and you can specify default values for parameters.

```python
def add(x, y=0):
    return x + y

result = add(3, 5)  # Arguments: 3, 5
print('Sum is:',result)
```

* **Return Value:**

Functions can optionally return a value using the return statement. If a function does not explicitly return a value, it returns None by default.

```python
def square(x):
    return x ** 2

result = square(4)  # Returns: 16
print('Square is:',result)
```

* **Docstrings:**

Docstrings are strings enclosed in triple quotes that provide documentation for functions. They describe what the function does, its parameters, and return values. Docstrings are accessible via the `__doc__` attribute of the function.

```python
def greet(name):
    """
    This function greets the user by name.
    """
    return f"Hello, {name}!"

print(greet.__doc__)
```
