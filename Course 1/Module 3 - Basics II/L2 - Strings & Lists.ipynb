## **2.1. Strings and some methods**

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Strings in Python are sequences of characters enclosed within single quotes (') or double quotes ("). They are immutable, meaning that once a string is created, it cannot be modified. Python provides a variety of built-in methods for manipulating and working with strings. Here are some commonly used string methods in Python:

* Conversion Methods:

`str.upper()`: Returns a copy of the string converted to uppercase.
`str.lower()`: Returns a copy of the string converted to lowercase.
`str.capitalize()`: Returns a copy of the string with the first character capitalized.
* Searching Methods:

`str.find(substring)`: Returns the lowest index in the string where the substring is found. Returns -1 if the substring is not found.
`str.index(substring)`: Returns the lowest index in the string where the substring is found. Raises a ValueError if the substring is not found.
* Counting Methods:

`str.count(substring)`: Returns the number of occurrences of the substring in the string.
* Replacement Methods:

`str.replace(old, new)`: Returns a copy of the string with all occurrences of the old substring replaced by the new substring.
`str.strip()`: Returns a copy of the string with leading and trailing whitespace removed.
`str.rstrip()`: Returns a copy of the string with trailing whitespace removed.
* Splitting and Joining Methods:

`str.split(separator)`: Splits the string into a list of substrings using the specified separator.
`str.join(iterable)`: Concatenates the elements of an iterable (such as a list) into a single string with the string as a separator.
* Checking Methods:

`str.startswith(prefix)`: Returns True if the string starts with the specified prefix; otherwise, returns False.
`str.endswith(suffix)`: Returns True if the string ends with the specified suffix; otherwise, returns False.
`str.isalpha()`: Returns True if all characters in the string are alphabetic; otherwise, returns False.
`str.isdigit()`: Returns True if all characters in the string are digits; otherwise, returns False.
`str.islower()`: Returns True if all alphabetic characters in the string are lowercase; otherwise, returns False.
`str.isupper()`: Returns True if all alphabetic characters in the string are uppercase; otherwise, returns False.

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;These are just some of the many string methods available in Python. String methods provide powerful tools for manipulating and working with text data in Python programs. You can explore more string methods in the Python documentation or by using the dir() function to inspect available methods for a string object.
