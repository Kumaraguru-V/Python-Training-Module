## 3.1. To print a string

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;In the previous lesson, we learnt how to input a string. Let's see how to output the string. In Python, output is called <b>'print'</b> because the output is printed on the screen/display. To print the string, we use another built-in function <b><i>print()</i></b>. In our analogy, the last step where we need to convey the successful registration to the user, we need to display the string, for which we use <b><i>print()</i></b>.

````python
print('Registration Successful! Welcome to the fair')
````

Now run the code cell below:

````python
'Registration Successful! Welcome to the fair'
````

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;You might get confused that, when you run the cell with a string, it gets printed, then why to explicitly use the <b><i>print()</i></b> function. The fact that we could see the strings from cells with just strings is a feature of Jupyter Notebook or simply notebook environment. After running a cell, Jupyter Notebook displays the content of the last line but not that of the previous lines. Run the cells below and see the difference. Also see the <b><i>print()</i></b> does not prints the quotes.</p>

````python
print('Registration Successful! Welcome to the fair')
print('You can get more information about the fair at contact points')
print('You can enjoy food and beverages at the food court area')
print('Scan the QR code for digital guide to the fair')
````

````python
'Registration Successful! Welcome to the fair'
'You can get more information about the fair at contact points'
'You can enjoy food and beverages at the food court area'
'Scan the QR code for digital guide to the fair'
````

Still confused? Here is an artistic way to show the difference. Run the cells below and see the difference.

````python
print (" /\_/\   ")
print (" >^.^<   ")
print ("  / \    ")
print (" (___)___")
````

Within a Jupyter Notebook cell, multiple lines of code can be written, with execution occurring sequentially. ## 3.2. Different use cases of <b><i>print()</i></b>

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<b><i>print()</i></b> can also be used to print various other information, like results of an arithmetic calculation. Afterall, the very first use case of programming language was to solve arithmetic calculations. Again Jupyter Notebook feature displays the last line of the cell even without print statement. Run the three code cells below to understand better.</p>

````python
# Case 1: The result of 5*7 will be printed
print(1)
print(2+3)
print(4*5)
print(6/2)
print((4*6)+7-(24/8))
5*7
````

````python
# Case 2: Only the result of last line, i.e. 5*7 will be printed
1
2+3
4*5
6/2
(4*6)+7-(24/8)
5*7
````

````python
# Case 3: Result of 5*7 will not be printed
print(1)
print(2+3)
print(4*5)
print(6/2)
print((4*6)+7-(24/8))
5*7
print(16*5)
````

In Case 3, notice that <b><i>5*7</i></b> did not print <b><i>35</i></b> in output, because neither there is an explicit print function nor it is the last line of the cell.

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<b><i>print()</i></b> and <b><i>input()</i></b> functions can be used within one another. Or most of the built-in functions can be used one within another provided it is logical. this is called nesting of functions. This also comes handy when writing complex codes which we'll learn in further modules. For now, run the cells below for better understanding.</p>

````python
# Case 1: Input nested within Print
print(input('Enter Name: '))
````

````python
# Case 2: Simple Input
input('Enter Name: ')
````

In case 1, Python directly prints the input given by the user. See the printed name is without quotes with print function.

````python
input(print('Enter your name below'))
````

In case 3, <b><i>print()</i></b> is nested within <b><i>input()</i></b>.This is logically possible, but doesn't make sense to the user.
