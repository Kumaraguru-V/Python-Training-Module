## 1.1. Syntax and Code Blocks
 <p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;When coding, ensuring proper syntax is crucial. Even the smallest error, such as a missing parentheses or an extra comma, can lead to a syntax error, causing the code to fail to execute. It's important to be vigilant, as errors can occur even with seemingly insignificant mistakes. If your code produces errors or exceptions, it's essential to carefully review the syntax and be mindful of minor errors.<br><br>&nbsp;&nbsp;&nbsp;Correct syntax alone may not guarantee desired outcomes; unexpected behavior or output could stem from semantic issues as well. Syntax governs the structure of code, while semantics dictate the meaning behind coded statements. It's vital to consider what the code aims to achieve. Although code may be syntactically correct and run without errors, it may not fulfill its intended purpose.<br><br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;When tackling code blocks in this course's exercises, be attentive to syntax errors and assess the overall functionality of your code. Simply rectifying a syntax error does not guarantee the desired outcome upon execution. After addressing any errors, ensure to submit your code for evaluation and verification.</p>
