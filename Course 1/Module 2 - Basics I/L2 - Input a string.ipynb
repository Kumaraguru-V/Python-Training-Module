## 2.1. Brief idea about function

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;We know from the previous lesson that strings can be used to represent questions and answers. Now how do we make the display to ask the question and accept the answer from the user?<br><br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;We use some special codewords for this purpose. They are called built-in functions. Now let's get a brief understanding of functions and built-in functions.</p>

<br>

**Function:**
<p align="justify"><I>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Function is a block of code which solves a problem or subset of a big problem. It can either be user-defined or built-in. For example in the above analogy, each step of the algorithm requires a certain lines of code. So we can define our functions like follows:</I></p>

* **ask_name** --> Ask the name
* **get_name** --> Accepts user input for name
* **ask_city** --> Ask the city
* **get_city** --> Accepts user input for city
* **print_info** --> Display greetings and registration success

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Note that in the above example, the functions are user-defined and the person who writes the code enjoys the freedom to name the fuctions as per his wish. <b>ask_name</b> can also be written as <b>ask_user_name</b> or anything as per your wish. And a function is always executed only when it is called to execute. We'll learn more about this in Module 3.

<br>
<br>

**Built-in function:**

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;For smooth functioning of the programming language, certain lines of code are pre-written in Python package by its creators. These are called built-in functions. And, when calling them, we have to use their names as defined in Python. Some built-in functions frequently used are <B><I>input(), print(), str(), int(), etc.</I></B><br>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;We can identify a built-in function in Jupyter Notebook by two distinct characteristics. Firstly, built-in functions are consistently highlighted in green. Secondly, they are always followed by parentheses '()', which we'll refer to as round brackets in this course to distinguish them from other types of brackets discussed later on. Within these round brackets, we frequently include an argument, such as a string containing the question we wish to ask, in the case of the input() function. Built-in functions are invaluable tools provided by the creators of a programming language to simplify coding tasks and enhance ease of use.</p>

## 2.2. Input a string

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;A built-in function called <B><I>input()</I></B> is used to accept a input from the user. What does this mean? Let's look at the code!</p>

```python
input('Enter your name:')
input('Enter your city:')
```
Now let's understand how this works. Run the code cell below.

````python
input('Enter your name: ')
````

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Several important events have occurred in this scenario! Initially, there is a number on the left side of the cell as anticipated. However, while responding to the question, instead of the number, a star symbol (*) appeared. This symbol indicates that the cell has commenced execution but has not yet completed. To finalize the cell run and execute the code, we need to press return or enter after providing the answer. Failure to complete the cell run prevents the code within the cell from executing, subsequently hindering the execution of subsequent cells. Now, shifting focus to the code itself, we observe that 'Enter your name: ' is recognized as a string due to being enclosed within quotes and displayed in red color. As for input(), this function enables us to prompt a user with a question. In Jupyter Notebook, invoking input() generates a text box (a white rectangle) where users can input text. Here, input() serves a specific function and is classified as a built-in function.<br><br>Now run the below cell to input city name.</p>

````python
input('Enter your city: ')
````
