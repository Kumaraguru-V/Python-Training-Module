## 3.1. What is Markdown?

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Jupyter notebook has two types of cells: Code and Markdown. Markdown cells are basically text editor where the text is formatted using a simple markup language called Markdown. To see the Markdown source, double-click a Markdown cell. When running the cell, jupyter displays the rendered version.</p>

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Now you might get a clarity as to what are those weird codings when you double-clicked the cells while reading the course. Yes, infact this entire course is developed in a notebook environment using Markdown</p>

## 3.2. Some useful cheatsheet for Markdown

<p align="justify">&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Some basic text formatting are done in Markdown as follows. You may play around yourselves by inserting a new row and trying out these cool tricks.</p>
<br>

Markdown | Preview
--- | ---
`**bold text**` | **bold text**
`*italicized text*` or `_italicized text_` | *italicized text*
`` `Monospace` `` | `Monospace`
`~~strikethrough~~` | ~~strikethrough~~
`[A link](https://www.google.com)` | [A link](https://www.google.com)
`![An image](https://www.google.com/images/rss.png)` | ![An image](https://www.google.com/images/rss.png)
